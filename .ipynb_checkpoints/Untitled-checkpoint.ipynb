```python
"""
1. Створіть xls-документ, що містить дані про марки автомобілів, їхню вартість, споживання палива, максимальну швидкість
та вантажопідйомність. Кількість записів у документі - 20.
2. Візуалізуйте залежності: споживання палива від вартості - лінійною діаграмою; споживання палива від швидкості - точковим
графіком; вантажопідйомність від швидкості - точковим графіком.
"""
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Завантаження даних із xls-файлу
data = pd.read_excel('Auto.xls')
print(data.head(20))
```

```python
# Споживання палива від вартості – лінійна діаграма
plt.plot(data['Вартість ($)'], data['Споживання палива'], 'bo')
plt.xlabel('Вартість ($)')
plt.ylabel('Споживання палива (л/100 км)')
plt.title('Залежність споживання палива від вартості')
plt.show()
```

```python
# Споживання палива від швидкості – точкова діаграма
plt.scatter(data['Максимальна швидкість'], data['Споживання палива'], c='red', marker='o')
plt.xlabel('Максимальна швидкість (км/г)')
plt.ylabel('Споживання палива (л/100 км)')
plt.title('Залежність споживання палива від швидкості')
plt.show()
```

```python
# Вантажопідйомність від швидкості – точкова діаграма
plt.scatter(data['Максимальна швидкість'], data['Вантажопідйомність'], c='green', marker='o')
plt.xlabel('Максимальна швидкість (км/г)')
plt.ylabel('Вантажопідйомність (кг)')
plt.title('Залежність вантажопідйомності від швидкості')
plt.show()
```

```python
"""
3. Побудуйте гістограму для розділення автомобілів на 5 групп за швидкістю.
"""
```

```python
# Розбиття даних на 5 груп за максимальною швидкістю
groups = pd.cut(data['Максимальна швидкість'], bins=5)
```

```python
# Побудова гістограми
plt.hist(data['Максимальна швидкість'], bins=5, color='green', alpha=0.5)
plt.xlabel('Максимальна швидкість (км/г)')
plt.ylabel('Число автомобілів')
plt.title('Поділ автомобілів на 5 груп за швидкістю')
plt.show()
```

```python
"""
4. Побудуйте коробковий графік для відображення швидкостей різних марок.
"""
```

```python
# Группування даних по марках автомобілів
groups = data.groupby('Марка')
```

```python
# Створення списку швидкостей для кожної марки
speeds = [group['Максимальна швидкість'].tolist() for name, group in groups]
```

```python
# Побудова коробкової діаграми
plt.boxplot(speeds)
plt.xticks(range(1, len(groups)+1), groups.groups.keys())
plt.xlabel('Марка автомобіля')
plt.ylabel('Максимальна швидкість (км/г)')
plt.title('Швидкості різних марок автомобілів')
plt.show()
```
